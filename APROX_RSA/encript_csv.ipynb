```python
fumar = []
with open('survey_lung_cancer.csv', 'r') as archivo: 
    lineas = archivo.read().splitlines()
    #borro encabezado
    lineas.pop(0)
    for l in lineas: 
        linea = l.split(',')
        fumar.append(int(linea[2]))
```

```python
fumar 
```

```python
e = 2**16 + 1
p = 409
q = 317 
n = p * q

phi_n = (p-1)*(q-1)

pubKey = [n, e]
priKey = [n, pow(e,-1,phi_n)]
```

```python
#creamos columna encriptada de los elementos fumar. En RSA el 1 no varia
fumar_encip = []
for elemento in fumar: 
    fumar_encip.append(E_RSA(pubKey, elemento))
```

```python
x = E_RSA(pubKey, 1)
y = E_RSA(pubKey, 2)
#fumador SI = 2, NO = 1
```

```python
from math import prod
producto = (prod(fumar_encip)) % n
producto
```

```python
valor = D_RSA(priKey, producto)
valor
#esta bien desencriptado porque da lo mismo que prod(fumar) % n
```
